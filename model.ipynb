```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/dataset.csv')
data.head()
```

```python
data["Symptoms"] = 0
```

```python
# collecting all symptoms of a sample and then placing it in another column of name "Symptoms" as list
records = data.shape[0]
for i in range(records):
    data["Symptoms"][i] = np.array([x.strip()
                           for x in data.iloc[i].values if type(x) == str])
```

```python
data["Symptoms"]
```

```python
column_values = data[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4',
       'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9',
       'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14',
       'Symptom_15', 'Symptom_16', 'Symptom_17']].values.ravel()
```

```python
# selecting unique symptoms from all columns of the dataframe

symps = pd.unique(column_values)
symps = symps.tolist()
print(len(symps)) # nan included
symps = [i.strip() for i in symps if type(i) == str]
print(len(symps)) # nan not included
```

```python
# now we making a dataframe with colum names as symptoms and shape of (4920, 131)
symptoms = pd.DataFrame(columns = symps,index = data.index)
```

```python
symptoms.head()
```

```python
# symptoms.head()
symptoms.shape
```

```python
symptoms["Symptoms"] = data["Symptoms"]
```

```python
symptoms.head()
```

```python
symps
```

```python
# now we are fillings the dataframe with appropiate data based on the symptoms 
# that can be used to train the model
for i in symps:
    symptoms[i] = symptoms.apply(lambda x:1 if i in x.Symptoms else 0, axis=1)
```

```python
symptoms.head()
```

```python
# attaching the Disease column
symptoms["Disease"] = data["Disease"]
# dropping the Symptoms column
symptoms = symptoms.drop("Symptoms",axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting the dataset to train and test
train, test = train_test_split(symptoms,test_size=0.2)
X_train = train.drop("Disease",axis=1)
y_train = train["Disease"].copy()
X_test = test.drop("Disease",axis=1)
y_test = test["Disease"].copy()
```

```python
# training on decision tree classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'entropy')
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred =  clf.predict(X_test)
```

```python
# we can see the accuracy results of the model
from sklearn.metrics import accuracy_score
print('Accuracy Score on train data: ', accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
print('Accuracy Score on test data: ', accuracy_score(y_true=y_test, y_pred=y_pred))
```
